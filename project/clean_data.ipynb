```python
import pandas as pd 
import numpy as np
```

### Let's Remove Redundant Columns

```python
sustain_w_regions = pd.read_csv("../data/sustainability_w_regions.csv")
print(len(sustain_w_regions.columns))

columns_to_remove = ["name", "alpha-2", "country-code", "World Regions (UN SDG Definition)", "region", "region-code", "iso_3166-2"]

cleaned_sustain_w_regions = sustain_w_regions.drop(columns=columns_to_remove)
print(len(cleaned_sustain_w_regions.columns))

cleaned_sustain_w_regions.to_csv("../data/cleaned_sustainability_w_regions.csv", index=False)
```

### Identifying Metadata

```python
cleaned_data = pd.read_csv("../data/cleaned_sustainability_w_regions.csv")

metadata_columns = ['Country Name', 'Country Code', 'Year', 'Continent', 'sub-region', 'intermediate-region', 'sub-region-code', 'intermediate-region-code']
data_columns = [col for col in cleaned_data.columns if col not in metadata_columns]

metadata_dict = {col: cleaned_data[col] for col in metadata_columns}
metadata = pd.DataFrame(metadata_dict)

data_dict = {col: cleaned_data[col] for col in data_columns}
data = pd.DataFrame(data_dict)

data = data.fillna(0)

data
```

### Converting to float values 

```python
column_headers = data.columns
np_data = np.array(data)
data = pd.DataFrame(np_data)
data.columns = column_headers

data
```

### Encode String entries

```python
string_cols = ["Income Classification (World Bank Definition)", "Regime Type (RoW Measure Definition)"]

for col in string_cols:
    cur_col = data[col]
    labels, uniques = pd.factorize(cur_col)
    data[col] = labels

data

data.to_csv("../data/sustainability_data.csv", index=False)
metadata.to_csv("../data/sustainability_metadata.csv", index=False)

```
